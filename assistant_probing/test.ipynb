## scenario_07_treatment_run00

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent))

from assistant_probing.projection import run_probe_multi, save_csv_multi

# --- configure paths here ---
AXIS_DIR = Path.cwd() / "data" / "axis"
ACTIVATIONS_DIR = Path.cwd().parent / "data" / "runs" / "scenario_07_treatment_run00" / "activations"
PROJECTION_DIR = Path.cwd() / "data" / "projections" / "scenario_07_treatment_run00"
RUN_NAME = "scenario_07_treatment_run00"
LAYERS = (32, 50)
# ----------------------------

scores = run_probe_multi(ACTIVATIONS_DIR, axis_dir=AXIS_DIR, layers=LAYERS)
for turn, (name, layer_scores) in enumerate(scores):
    cols = "  ".join(f"L{layer} = {layer_scores[layer]: 9.4f}" for layer in LAYERS)
    print(f"turn {turn:>2}  {cols}   {name}")

out_path = save_csv_multi(scores, RUN_NAME, output_dir=PROJECTION_DIR, layers=LAYERS)
print(f"saved to {out_path}")
```
