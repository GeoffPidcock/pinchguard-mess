## Load pools

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd()))

import generate_scenarios as gs

baseline_pool, treatment_pool = gs.load_pools()
print(f"baseline pool: {len(baseline_pool)}  treatment pool: {len(treatment_pool)}")
```

## Generate scenarios

`generate_scenarios.py --n` defaults to 1; this notebook runs 5.

```python
import random

N = 5

used_ids_path = Path.cwd() / "used_ids.json"
output_dir = (Path.cwd() / "../../scenarios/099 auto-generated scenarios").resolve()
output_dir.mkdir(parents=True, exist_ok=True)

used_ids = gs.load_used_ids(used_ids_path)
rng = random.Random()

for _ in range(N):
    gs.generate_scenario(baseline_pool, treatment_pool, used_ids, output_dir, rng)
    gs.save_used_ids(used_ids_path, used_ids)
```
